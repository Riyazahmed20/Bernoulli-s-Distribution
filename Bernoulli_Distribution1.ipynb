```python
import matplotlib.pyplot as plt 
from scipy.stats import bernoulli

#Instance of Bernoulli distribution with porometer p= 0.6 
bd=bernoulli(0.6)

#Outcome of experiment can take value as 0, 1 
X= [0, 1]

plt.figure(figsize=(6,6))
plt.xlim(-1, 2)

#Create a bor plot 
plt.bar(X, bd.pmf(X), color='red') 
plt.title('Bernoulli Distribution (p=0.6)', fontsize='12')

plt.xlabel('Values of Random Variable X (0,1)', fontsize='12')

plt.ylabel('Probability',fontsize='12')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#Creating a series of data of in range of 1-50
x=np.linspace(1,50,100)
#Creating a Function
def normal_dist(x,mean,sd):
    prob_density=(np.pi*sd)*np.exp(-0.5*((x-mean)/sd)**2)
    return prob_density
#Calculate mean and Standard deviation
mean=np.mean(x)
sd=np.std(x)

#Apply function to the data.
pdf=normal_dist(x,mean,sd)

#Plotting the Results
plt.plot(x,pdf, color='red')
plt.xlabel('Data points')
plt.ylabel('Probability Density')
```
